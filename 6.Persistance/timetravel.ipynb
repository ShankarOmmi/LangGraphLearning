```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.checkpoint.memory import InMemorySaver
from dotenv import load_dotenv



load_dotenv()

llm = ChatGoogleGenerativeAI(model ='gemini-2.5-flash')
```

```python
class QuoteState(TypedDict):

    topic : str
    quote : str
    explanation : str
```

```python
def generate_quote(state: QuoteState):

    prompt = f'generate a single short yet meaningful quote on the topic {state["topic"]}'
    response = llm.invoke(prompt).content

    return {'quote': response}
```

```python
def generate_explanation(state: QuoteState):

    prompt = f'write an explanation for the quote - {state["quote"]}'
    response = llm.invoke(prompt).content

    return {'explanation': response}
```

```python
graph = StateGraph(QuoteState)

graph.add_node('generate_quote', generate_quote)
graph.add_node('generate_explanation', generate_explanation)

graph.add_edge(START, 'generate_quote')
graph.add_edge('generate_quote', 'generate_explanation')
graph.add_edge('generate_explanation', END)

checkpointer = InMemorySaver()

workflow = graph.compile(checkpointer=checkpointer)
```

```python
config1 = {"configurable": {"thread_id": "1"}}
workflow.invoke({'topic':'Life'}, config=config1)
```

```python
workflow.get_state(config1)
```
